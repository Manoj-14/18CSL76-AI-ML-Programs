```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6,4,1000)
x = x  + np.random.normal(scale = 0.05 , size =1000)
y = np.sin(x)

plt.scatter(x,y,alpha=0.3)
```

```python
def radial_kernel(x0,x,tau):
    return np.exp(np.sum((x-x0) ** 2 , axis = 1)/(-2 * tau ** 2))
```

```python
def local_reg(x0,x,y,tau):
    x0 = np.r_[1,x0]
    x = np.c_[np.ones(len(x)),x]
    xw = x.T * radial_kernel(x0,x,tau)
    beta = np.linalg.pinv(xw @ x) @ xw @ y
    return x0 @ beta
```

```python
def plot_lr(tau):
    domine = np.linspace(-6,4,300)
    pred =[local_reg(x0,x,y,tau) for x0 in domine]
    plt.scatter(x,y,alpha=0.3)
    plt.plot(domine,pred,c="r")
    return plt
```

```python
plot_lr(0.05).show()
```
