```python
class Graph :

    def __init__(self,adj_graph,heuristic , start) :
        self.adj_graph = adj_graph
        self.heuristic = heuristic
        self.start = start
        self.status = {}
        self.par = {}
        self.solution ={}

    def get_neig(self,v):
        return self.adj_graph.get(v,"")

    def get_status(self,v):
        return self.status.get(v,0)

    def set_status(self,v , val):
        self.status[v] = val

    def get_h(self,v):
        return self.heuristic.get(v,0)

    def set_h(self,v,val):
        self.heuristic[v] = val

    def print_sol(self):
        print("Solution:",self.solution)

    def apply_ao_star(self):
        self.ao_star(self.start , False)

    def conMinNode(self,v):
        mincost = 0
        cost_of_child_dict = {mincost:[]}
        flag = True

        for nodes in self.get_neig(v):
            cost = 0
            nodeList = []
            for node , weight in nodes :
                cost = cost + weight + self.get_h(node)
                nodeList.append(node)

            if flag == True :
                mincost = cost
                cost_of_child_dict[mincost] = nodeList
                flag = False

            else:
                if mincost > cost :
                    mincost = cost
                    cost_of_child_dict[mincost] = nodeList
        print(cost_of_child_dict)
        return mincost , cost_of_child_dict[mincost]

    def ao_star(self,v,back_track):
        print("-"*30)
        print("Heuristic Value:",self.heuristic)
        print("Solution:",self.solution)
        print("Processing node:",v)
        print("-"*30)

        if self.get_status(v) >= 0:
            mincost , nodeList = self.conMinNode(v)
            print("Min cost",mincost)
            print("Nodelist",nodeList)
            self.set_h(v,mincost)
            self.set_status(v,len(nodeList))
            solved = True

            for child in nodeList :
                self.par[child] = v
                print("child",child,self.get_status(child))
                if self.get_status(child) !=-1 :
                    solved = False
            
            if solved is True :
                self.set_status(v , -1)
                self.solution[v] =  nodeList

            if v != self.start :
                print("Backtrack True")
                self.ao_star(self.par[v],True)
                print("--")
            
            if back_track is False :
                print("Backtrack False")
                print("nodelist",nodeList)
                for node in nodeList :
                    self.set_status(node,0)
                    self.ao_star(node , False)


```

```python
h1 = {'A': 1, 'B': 3, 'C': 4, 'D': 5, 'E': 5, 'F': 7, 'G': 4, 'H': 4}
graph1 = {
'A': [[('B', 1), ('C', 1)], [('C', 1),('D', 1)]],
'B': [[('E', 1)], [('F', 1)]],
'D': [[('G', 1), ('H', 1)]],
}
```

```python
graph = Graph(graph1,h1,"A")
graph.apply_ao_star()
```
