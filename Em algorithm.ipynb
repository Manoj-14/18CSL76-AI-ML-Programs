```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris ,load_wine
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
data = pd.DataFrame(iris.data,columns = iris.feature_names)
y = iris.target
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.scatter(data["sepal length (cm)"],data["sepal width (cm)"],c = y, s=40)
plt.title("Sepel")
plt.subplot(1,2,2)
plt.scatter(data["petal length (cm)"],data["petal width (cm)"],c = y, s=40)
plt.title("Petal")
```

```python
k_means = KMeans(n_clusters=2)
k_means.fit(data )
k_means_pred = k_means.predict(data)
accuracy_score(y,k_means_pred)
```

```python
gm = GaussianMixture(n_components = 3)
gm.fit(data)
gm_pred = gm.predict(data)
accuracy_score(y,gm_pred)
```

```python
plt.figure(figsize = (20,7))
plt.subplot(1,3,1)
plt.title("Actual classes")
plt.scatter(data["petal length (cm)"],data["petal width (cm)"],c = y, s=40)

plt.subplot(1,3,2)
plt.title("KMeans")
plt.scatter(data["petal length (cm)"],data["petal width (cm)"],c = k_means_pred, s=40)

plt.subplot(1,3,3)
plt.title("Em cluster")
plt.scatter(data["petal length (cm)"],data["petal width (cm)"],c = gm_pred, s=40)
```

```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
print(iris.feature_names)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.title("Sepal")
plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=y,s=40)

plt.subplot(1,2,2)
plt.title("Petal")
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=y,s=40)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
kmeans_predict = kmeans.predict(x)

print("accuracy of kmeans:",accuracy_score(y,kmeans_predict))
```

```python
gm = GaussianMixture(n_components=3)
gm.fit(x)
gm_predict = gm.predict(x)

print("accuracy of EM:",accuracy_score(y,gm_predict))
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1,3,1)
plt.title("Actual Class")
plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=y,s=40)

plt.subplot(1,3,2)
plt.title("Kmeans")
plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=kmeans_predict,s=40)

plt.subplot(1,3,3)
plt.title("EM")
plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=gm_predict,s=40)
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1,3,1)
plt.title("Actual Class")
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=y,s=40)

plt.subplot(1,3,2)
plt.title("Kmeans")
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=kmeans_predict,s=40)

plt.subplot(1,3,3)
plt.title("EM")
plt.scatter(x['petal length (cm)'],x['petal width (cm)'],c=gm_predict,s=40)
```
